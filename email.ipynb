```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('combined_lemm_stop.csv')
data.head()
```

```python
data.content[0]
```

```python
print(data.content.str.split('\n')[0])
```

```python
s = data.content.str.split('\n').apply(pd.Series)
data['subject'] = s[0]
data['msg'] = s[2]
data.head()
```

```python
data.label = [0 if 'legit' in file else 1 for file in data.filename]
data.head()
```

```python
data.subject = data.subject.str.replace('Subject: ', '')
data.head()
```

```python
data.drop(columns=['folder', 'filename', 'content'], inplace=True)
```

```python
data['combined'] = data.subject + ' ' + data.msg
```

```python
data.head()
```

```python
from sklearn.feature_selection import mutual_info_classif, SelectKBest
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer()
data.combined=cv.fit_transform(data.combined)
```
